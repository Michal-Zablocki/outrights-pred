```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel('data/sheets/optaelo20240524.xlsx')
```

```python
df = df.sort_values(by='Club elo')
df = df.iloc[1:-1, :]
```

```python
X = df['Opta power ranking']
y = df['Club elo']
```

```python
lr = LinearRegression()
lr.fit(X.values.reshape(-1, 1), y.values)
```

```python
lr.score(X.values.reshape(-1, 1), y.values)
```

```python
print(f'y = {lr.coef_[0]} * x + {lr.intercept_}')
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X.values.reshape(-1, 1)), color='red')
```
